```python

import os
import shutil

# Define the number of classes and their labels
nc = 7
classes = ['elbow positive', 'fingers positive', 'forearm fracture', 'humerus fracture', 'humerus', 'shoulder fracture', 'wrist positive']

# Function to organize images into class folders based on their label files
def organize_images_by_labels(label_folder, image_folder, output_folder):
    # List all label files in the label folder
    label_files = [f for f in os.listdir(label_folder) if f.endswith('.txt')]

    for label_file in label_files:
        # Extract the image name from the label file (remove the .txt extension)
        image_name = label_file.replace('.txt', '')

        # Open the label file and extract class information
        label_path = os.path.join(label_folder, label_file)

        with open(label_path, 'r') as file:
            lines = file.readlines()

        # Check if the label file is not empty
        if not lines:
            print(f"Label file {label_file} is empty or improperly formatted.")
            continue  # Skip to the next label file if this one is empty or invalid

        # Assume the class index is the first number in the label file
        try:
            class_index = int(lines[0].strip().split()[0])
            class_name = classes[class_index]
        except (IndexError, ValueError) as e:
            print(f"Error processing {label_file}: {e}")
            continue

        # Create the output class folder if it doesn't exist
        class_folder = os.path.join(output_folder, class_name)
        os.makedirs(class_folder, exist_ok=True)

        # Find the corresponding image file in the image folder (assuming JPG format)
        image_path = os.path.join(image_folder, image_name + ".jpg")

        # Check if the image file exists, then copy it to the class folder
        if os.path.exists(image_path):
            shutil.copy(image_path, os.path.join(class_folder, image_name + ".jpg"))
        else:
            print(f"Image {image_path} not found!")
```

```python
# Define paths for train label folder, image folder, and output folder
train_label_folder = os.path.join(Train_dir, 'labels')  # Folder where label .txt files are stored
train_image_folder = os.path.join(Train_dir, 'images')  # Folder where the images are stored
train_output_folder = '/content/drive/MyDrive/bone fracture detection1/train_output'  # Folder where organized images will be stored

# Call the function to organize images based on label files
organize_images_by_labels(
    label_folder=train_label_folder,
    image_folder=train_image_folder,
    output_folder=train_output_folder
)
```

```python
import torch
import torchvision
import tarfile
import torch.nn as nn
import torch.nn.functional as F
from torchvision.datasets.utils import download_url
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader
import torchvision.transforms as tt
from torchvision.utils import make_grid
from torchvision.transforms import ToTensor
```

```python


from torch.utils.data import Dataset, DataLoader
import torchvision.transforms as transforms

batch_size=64
transform = transforms.Compose([tt.Resize((256, 256)),
                        tt.RandomRotation(0),
                        tt.ToTensor()])


dTrain = ImageFolder(train_output_folder, transform = transform)
dTest = ImageFolder(test_output_folder, transform = transform)

train_loader = DataLoader(dTrain, batch_size, shuffle=True,num_workers=4)
test_loader = DataLoader(dTest, batch_size, shuffle=False,num_workers=4)
```

```python

def get_default_device():
    """Pick GPU if available, else CPU"""
    if torch.cuda.is_available():
        return torch.device('cuda')
    else:
        return torch.device('cpu')

def to_device(data, device):
    """Move tensor(s) to chosen device"""
    if isinstance(data, (list,tuple)):
        return [to_device(x, device) for x in data]
    return data.to(device, non_blocking=True)

class DeviceDataLoader():
    """Wrap a dataloader to move data to a device"""
    def __init__(self, dl, device):
        self.dl = dl
        self.device = device

    def __iter__(self):
        """Yield a batch of data after moving it to device"""
        for b in self.dl:
            yield to_device(b, self.device)

    def __len__(self):
        """Number of batches"""
        return len(self.dl)
```

```python
device = get_default_device()
device
```

```python
train_dl = DeviceDataLoader(train_loader, device)
test_dl = DeviceDataLoader(test_loader, device)
```

```python
def accuracy(outputs, labels):
    _, preds = torch.max(outputs, dim=1)
    return torch.tensor(torch.sum(preds == labels).item() / len(preds))

class ImageClassificationBase(nn.Module):
    def training_step(self, batch):
        images, labels = batch
        out = self(images)                  # Generate predictions
        loss = F.cross_entropy(out, labels) # Calculate loss
        return loss

    def validation_step(self, batch):
        images, labels = batch
        out = self(images)                    # Generate predictions
        loss = F.cross_entropy(out, labels)   # Calculate loss
        acc = accuracy(out, labels)           # Calculate accuracy
        return {'val_loss': loss.detach(), 'val_acc': acc}

    def validation_epoch_end(self, outputs):
        batch_losses = [x['val_loss'] for x in outputs]
        epoch_loss = torch.stack(batch_losses).mean()   # Combine losses
        batch_accs = [x['val_acc'] for x in outputs]
        epoch_acc = torch.stack(batch_accs).mean()      # Combine accuracies
        return {'val_loss': epoch_loss.item(), 'val_acc': epoch_acc.item()}

    def epoch_end(self, epoch, result):
        print("Epoch [{}], last_lr: {:.5f}, train_loss: {:.4f}, val_loss: {:.4f}, val_acc: {:.4f}".format(
            epoch, result['lrs'][-1], result['train_loss'], result['val_loss'], result['val_acc']))
```

```python
from torchvision import models

class BoneFractureModel(ImageClassificationBase):
    def __init__(self, num_classes, pretrained=True):
        super().__init__()
        # Use a pretrained model
        self.network = models.resnet34(pretrained=pretrained)
        # Replace last layer
        self.network.fc = nn.Linear(self.network.fc.in_features, num_classes)

    def forward(self, xb):
        return self.network(xb)
```

```python

import torch
from tqdm.notebook import tqdm

@torch.no_grad()
def evaluate(model, val_loader):
    model.eval()
    outputs = [model.validation_step(batch) for batch in val_loader]
    return model.validation_epoch_end(outputs)


def fit(epochs, lr, model, train_loader, val_loader, opt_func=torch.optim.SGD):
    history = []
    optimizer = opt_func(model.parameters(), lr)
    for epoch in range(epochs):
        # Training Phase
        model.train()
        train_losses = []
        for batch in tqdm(train_loader):
            loss = model.training_step(batch)
            train_losses.append(loss)
            loss.backward()
            optimizer.step()
            optimizer.zero_grad()
        # Validation phase
        result = evaluate(model, val_loader)
        result['train_loss'] = torch.stack(train_losses).mean().item()
        model.epoch_end(epoch, result)
        history.append(result)
    return history

def get_lr(optimizer):
    for param_group in optimizer.param_groups:
        return param_group['lr']

def fit_one_cycle(epochs, max_lr, model, train_loader, val_loader,
                  weight_decay=0, grad_clip=None, opt_func=torch.optim.SGD):
    torch.cuda.empty_cache()
    history = []

    # Set up custom optimizer with weight decay
    optimizer = opt_func(model.parameters(), max_lr, weight_decay=weight_decay)
    # Set up one-cycle learning rate scheduler
    sched = torch.optim.lr_scheduler.OneCycleLR(optimizer, max_lr, epochs=epochs,
                                                steps_per_epoch=len(train_loader))

    for epoch in range(epochs):
        # Training Phase
        model.train()
        train_losses = []
        lrs = []
        for batch in tqdm(train_loader):
            loss = model.training_step(batch)
            train_losses.append(loss)
            loss.backward()

            # Gradient clipping
            if grad_clip:
                nn.utils.clip_grad_value_(model.parameters(), grad_clip)

            optimizer.step()
            optimizer.zero_grad()

            # Record & update learning rate
            lrs.append(get_lr(optimizer))
            sched.step()

        # Validation phase
        result = evaluate(model, val_loader)
        result['train_loss'] = torch.stack(train_losses).mean().item()
        result['lrs'] = lrs
        model.epoch_end(epoch, result)
        history.append(result)
    return history
```

```python
model = BoneFractureModel(len(dTrain.classes))
to_device(model, device);
```
